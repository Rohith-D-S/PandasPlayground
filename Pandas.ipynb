```python
# When we filter the data, the old index remains as is. Inorder to fix this
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2'] =='Poison') & (df['HP']>70)] 
new_df =new_df.reset_index(drop=True) #Bydefault it saves old index to new column so inorder to drop that column pass drop=True as an argument
new_df
```

```python
#if you dont want to assign to new data frame you can do it inplace
new_df=df.loc[(df['Type 1']=='Grass') & (df['Type 2'] =='Poison') & (df['HP']>70)] 
new_df.reset_index(drop=True,inplace=True) #Bydefault it saves old index to new column so inorder to drop that column pass drop=True as an argument
new_df
```

# AGGREGATE STATISTICS (GROUP BY)

```python
1.mean()
2.sum()
3.count()
  etc..
```

```python
# calculate the mean of all numeric columns grouped by 'Type 1'
df=pd.read_csv('modified.csv')
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby(['Type 1'])[numeric_cols].mean()

df.head(4)
```

```python
#To see the Highest defense
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby(['Type 1'])[numeric_cols].mean().sort_values('Defense',ascending=False) #To get the highest attack replace Defense with Attack
df.head(3)
```

```python
#sum function 
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby(['Type 1'])[numeric_cols].sum()
df.head(2)
```

```python
#count function
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby(['Type 1'])[numeric_cols].count()
df.head(2)
```

```python
#Add count column to DataFrame
df['count']=1
df.head(2)
```

```python
#count function with multiple parameter
numeric_cols = df.select_dtypes(include=['number']).columns
df.groupby(['Type 1','Type 2'])[numeric_cols].count()['count']
df.head(3)
```

# WORKING WITH LARGE AMOUNT OF DATA

```python
#When we are dealing with large amount of data instead of reading everything at a time we can read in chunks to save memory
#Using chunksize parameter
count=0
for df in pd.read_csv('modified.csv',chunksize=5):
    count+=1
    print("Chunk DataFrame:")
    print(df.head(5))
    if count>1:
        break
```

```python
new_df=pd.DataFrame(columns=df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
    results=df.groupby(['Type 1']).count()
    new_df=pd.concat([new_df,results])
    
```
